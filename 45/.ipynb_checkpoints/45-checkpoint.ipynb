```python
import pandas as pd
import geopy
from geopy.geocoders import Nominatim, Yandex
```

```python
df = pd.read_csv('data-20160705-structure-20150320.csv', sep=";", encoding='utf-8')
```

```python
df.head()
```

```python
geolocator = Yandex()
```

```python
import time
```

```python
def geo(x):
    print(x)
    time.sleep(1)
    location = geolocator.geocode(x)
    print((location.latitude, location.longitude))
    return pd.Series({'lat': location.latitude, 'lon': location.longitude})
```

```python
latlons = df['address'].apply(geo)
```

```python
def join(x):
    x.notnull
df['contacts'] = df[['email', 'fax', 'telephone', 'telephone1', 'url']].apply(lambda x: '; '.join(x[x.notnull()]), axis=1)
```

```python
customs = pd.concat([df['contacts'], latlons], axis=1)
```

```python
customs.to_csv('customs.csv', encoding='utf-8')
```
